```python
import cv2
```

```python
def select_tracker_api():
    api = input("Select the API to use for tracking, press\n\
                   0 for BOOSTING\n\
                   1 for MIL\n\
                   2 for KCF\n \
                   3 for TLD\n \
                   4 for MEDIANFLOW\n \
                   5 for GOTURN\n \
                   6 for MOSSE\n \
                   7 for CSRT: ")
    if api=='0':
        tracker_api = cv2.TrackerBoosting_create()
    """Boosting Tracker
    This is a real-time object tracking based on a novel on-line 
    version of the AdaBoost algorithm. The classifier uses surrounding
    background as negative examples in update step to avoid drifting 
    problem. """    
        
    elif api=='1':
        tracker_api = cv2.TrackerMIL_create()
    """The MIL algorithm trains a classifier in an online manner to
    separate the object from the background.
    Multiple Instance Learning avoids drift problem for robust tracking
    . """    
        
        
    elif api=='2':
        tracker_api = cv2.TrackerKCF_create()
    """the KCF (Kernelized Correlation Filter) tracker KCF is novel 
    tracking framework that utilizes properties of circulant matrix to
    enhance the processing speed. """    
        
        
    elif api=='3':
        tracker_api = cv2.TrackerTLD_create()
    """TLD (Tracking, learning and detection) tracker is novel tracking
    framework that explicitly decomposes the long-term tracking task 
    into tracking, learning and detection.
    The tracker follows the object from frame to frame. The detector 
    localizes all appearances that have been observed so far and 
    corrects the tracker if necessary. The learning estimates 
    detector's errors and updates it to avoid these errors in future. """
    
    
    elif api=='4':
        tracker_api = cv2.TrackerMedianFlow_create()
    """The tracker is suitable for very smooth and predictable movements 
    when object is visible throughout the whole sequence. It's quite and
    accurate for this type of problems (in particular, it was shown by
    authors to outperform MIL)."""    
        
    elif api=='5':
        tracker_api = cv2.TrackerGOTURN_create()
    """GOTURN (Generic Object Tracking Using Regression Networks) 
    tracker is kind of trackers based on Convolutional Neural Networks
    (CNN). While taking all advantages of CNN trackers, GOTURN is much
    faster due to offline training without online fine-tuning nature. 
    GOTURN tracker addresses the problem of single target tracking: 
    given a bounding box label of object in the first frame of video,
    we track that object through the rest of the video. NOTE: Current
    method of GOTURN does not handle occlusions; however, it is fairly
    robust to viewpoint changes, lighting changes, and deformations. 
    Inputs of GOTURN are two RGB patches representing Target and 
    Search patches resized to 227x227. Outputs of GOTURN are predicted
    bounding box coordinates, relative to Search patch coordinate 
    system, in format X1,Y1,X2,Y2. """
    
    elif api=='6':
        tracker_api = cv2.TrackerMOSSE_create()
    """MOSSE (Minimum Output Sum of Squared Error) tracker: The 
    implementation is based on [21] Visual Object Tracking using 
    Adaptive Correlation Filters. Note: this tracker works with
    grayscale images, if passed bgr ones, they will get converted
    internally."""    
        
        
    elif api=='7':
        tracker_api = cv2.TrackerCSRT_create()
    """The implementation is based on Discriminative Correlation Filter
    with Channel and Spatial Reliability"""    
    return tracker_api
                   
```

```python
tracker = select_tracker_api()
tracker_name = str(tracker).split()[0][1:]

# cap = cv2.VideoCapture(0)
cap = cv2.VideoCapture('videos/face.mp4')

ret, frame = cap.read()
roi = cv2.selectROI(frame, False)
"""Selects ROI on given image. Function creates window and allows user
to select ROI using mouse. Controls: use space or enter to finish 
selection, use key c to cancel selection """

ret = tracker.init(frame, roi)
"Initialize tracker with known bounding box that surrounded target."

while True:
    try:
        ret, frame = cap.read()

        success, roi = tracker.update(frame)
        "Update tracker, find new most likely bounding box for target."
        (x,y,w,h) = tuple(map(int, roi))

        if success:
            pts1 = (x,y)
            pts2 = (x+w, y+h)
            cv2.rectangle(frame, pts1, pts2, (255,125,25),3)

        else:
            cv2.putText(frame, 'Could not track the object!',
                       (100,200), cv2.FONT_HERSHEY_SIMPLEX,
                       1,(255,120,100),3)

        cv2.putText(frame, tracker_name,
                       (20,400), cv2.FONT_HERSHEY_SIMPLEX,
                       1,(255,10,200),3)
        try:
            cv2.imshow(tracker_name, frame)
        except:
            cap.release()
            cv2.destroyAllWindows()
            break        

        if cv2.waitKey(50) & 0xFF==27:
            break
    except:
        cap.release()
        cv2.destroyAllWindows()
        

cap.release()
cv2.destroyAllWindows()

        
        
```
