```python
import cv2
import numpy as np

cap = cv2.VideoCapture('videos/demo1.mp4')

ret, first_frame = cap.read()
print(first_frame, type(first_frame))
print(first_frame.shape)

prev_gray = cv2.cvtColor(first_frame, cv2.COLOR_BGR2GRAY)

mask = np.zeros(first_frame)


mask[..., 1] = 255

while(cap.isOpened()):
    ret, frame = cap.read()
    cv2.imshow('input', frame)
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    flow = cv2.calcOpticalFlowFarneback(prev_gray,
                                       gray,
                                       None,
                                       0.5,
                                       3,
                                       15,
                                       3, 
                                       5,
                                       1.2,
                                       0)
    magn, angle = cv2.cartToPolar(flow[..., 0],
                                 flow[..., 1])
    
    # Set image hue depending on optical flow direction
    mask[.../ 0] = angle*180/np.pi/2
    
    # Normalize magnitude
    mask[..., 2] = cv2.normalize(magn,
                                None,
                                0,
                                255,
                                cv2.NORM_MINMAX)
    
    # Convert hsv to rgb
    rgb = cv2.Color(mask, cv2.COLOR_HSV2RGB)
    
    cv2.imshow('Dense Optical Flow', rgb)
    prev_gray = gray
    
    if cv2.waitKey(0) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
    
```
